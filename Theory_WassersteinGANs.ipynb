### PROBLEMS WITH THE SIMPLE GAN:

1. **MODE COLLAPSE:**
  Generator stuck in a single mode (peak of distribution).

  Lets take an example of the simple GAN model created on the MNIST dataset, each number on the data distribution has a different peak which we call as modes. What may happen sometimes, when the balance of the training  is not good enough,the generator might discover that it can fool the discriminator by generating a certain mode or a number and begin to continuosly generate that. In this case, it may generate a lot of 1s because it is the easiest to learn and generate.


2. **LOSS VALUE MEANING:**
  Not a good correlation between the loss value and the quality of the generated results.

  The loss value dont really emphasize on the quality of results generated which is because of the nature of the min-max battle between the generator and the discriminator in these basic GANs.
